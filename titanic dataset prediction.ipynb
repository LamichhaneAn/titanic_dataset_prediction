Survived: Indicates whether the passenger

Sibsp: Number of siblings/spouse abroad the Titanic.
Parch: Number of parents/children abroad the Titanic.
Ticket: Ticket number.

```python
#Fill Missing Values

#We can fill the missing values in the Age column with the mean age and for Embarked use the mode.

#Age is a numerical value so .mean()
#Embarked is a Object Value so .mode()
```

```python
# Fill missing values in "Age" with the mean age
df["Age"].fillna(df["Age"].mean(), inplace=True)

# Fill missing values in "Embarked" with the mode
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
#Age Distribution:

import matplotlib.pyplot as plt

plt.hist(df["Age"], bins = 20, edgecolor = "red")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Survival Rate by Age:

# Define age bins and labels
age_bins = [0, 12, 20, 40, 60, 80]
age_labels = ["0-12", "13-20", "21-40", "41-60", "61-80"]

# Create AgeGroup column
df["AgeGroup"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels)

# Create survival rate by age group
survival_rate_by_age = df.groupby("AgeGroup")["Survived"].mean()

# Print survival rate by age group
print("Survival rate by age group:")
print(survival_rate_by_age * 100)
```

```python
# 5. Visualizatio: lets create the Visualization for better understanding the data

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate survival rate by gender
survival_rate_by_gender = df.groupby("Sex")["Survived"].mean()

# Print survival rate by gender
for gender, rate in survival_rate_by_gender.items():
    print(f"Survival rate for gender {gender}: {rate * 100:.2f}%")

# Plot survival rate by gender
sns.barplot(x=survival_rate_by_gender.index, y=survival_rate_by_gender.values)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#Survival Rate by Class:

#Calculate survival rate by class:
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate survival rate by class
survival_rate_by_class = df.groupby("Pclass")["Survived"].mean()

# Print survival rate by gender
for Pclass, rate in survival_rate_by_class.items():
    print(f"Survival rate for  {Pclass}: {rate * 100:.2f}%")

sns.barplot(x="Pclass", y = "Survived", data=df)
plt.title("Survival Rate by Class")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the Titanic dataset
df = pd.read_csv("titanic.csv")

# Data Processing
df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"C": 0, "Q": 1, "S": 2})
df.fillna(df.mean(), inplace=True)

# Split the dataset into features and target
X = df.drop("Survived", axis=1)
y = df["Survived"]

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build a machine learning model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```
